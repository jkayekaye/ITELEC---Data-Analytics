```python
import numpy as np
import pandas as pd 
import seaborn as sns 
import matplotlib.pyplot as plt 
import warnings
warnings.filterwarnings('ignore')
```

```python
dengue = pd.read_csv('datasets\\dengue.csv')
```

```python
dengue['Month'] = dengue['Month'].astype('category')
dengue['Region'] = dengue['Region'].astype('category')
```

<h1>1. How did dengue cases change over the years?</h1>

```python
plt.figure(figsize=(8,5))
sns.lineplot(x='Year', y='Dengue_Cases', data=dengue, marker='o')
plt.title('Trend of Dengue Cases by Year')
plt.show()
```

<h2>Insight No.1</h2>
<h3>The line shows all of the years the dengue cases increased, decreased, or fluctuated over time</h3>

<h2>Insight No.2</h2>
<h3>Based on the graph, the number of dengue cases went up and down over the years shown. Dengue cases were at their highest in 2019 and then dropped sharply in 2020. I think it's because of the time that the Philippines experienced COVID-19 pandemic.</h3>

<h1>2. Do dengue deaths follow the same pattern as dengue cases?</h1>

```python
# Visualization 2: Line plot of Dengue Deaths by Year
plt.figure(figsize=(8,5))
sns.lineplot(x='Year', y='Dengue_Deaths', data=dengue, marker='o', color='red')
plt.title('Trend of Dengue Deaths by Year')
plt.show()
```

<h2>Insight No.3</h2>
<h3>The data shows that dengue deaths don't match the pattern of dengue cases. Also dengue deaths have their own trends. There was a big drop in deaths from 2016 to 2018, followed by a slight increase and then another decrease. </h3>

<h1>3.Which region reports the highest and lowest average dengue cases?</h1>

```python
# Visualization 3: Bar chart of Average Dengue Cases by Region
plt.figure(figsize=(8,5))
sns.barplot(x='Region', y='Dengue_Cases', data=dengue, estimator='mean', ci=None, palette='viridis')
plt.title('Average Dengue Cases per Region')
plt.xticks(rotation=45)
plt.show()
```

<h2>Insight No.4</h2>
<h3>Based on the bar chart data, the region with the highest average dengue cases is Region III.</h3>

<h2>Insight No.5</h2>
<h3>Based on the bar chart titled data the region with the highest average dengue cases is Region BARMM(Bangsamoro Autonomous Region in Muslim Mindanao).</h3>

<h1>4. Which region has the highest and lowest dengue-related deaths?</h1>

```python
# Visualization 4: Bar chart of Average Dengue Deaths by Region
plt.figure(figsize=(8,5))
sns.barplot(x='Region', y='Dengue_Deaths', data=dengue, estimator='mean', ci=None, palette='coolwarm')
plt.title('Average Dengue Deaths per Region')
plt.xticks(rotation=45)
plt.show()
```

<h2>Insight No.6</h2>
<h3>Based on the bar chart, the region with the highest average dengue-related deaths is CAR</h3>

<h2>Insight No.7</h2>
<h3>Based on the bar chart, the region with the region with the lowest is NCR.</h3>

<h1>5. During which months are dengue cases highest?</h1>

```python
# Visualization 5: Boxplot of Dengue Cases by Month
plt.figure(figsize=(8,5))
sns.boxplot(x='Month', y='Dengue_Cases', data=dengue, palette='Set2')
plt.title('Dengue Cases Distribution by Month')
plt.xticks(rotation=45)
plt.show()
```

<h2>Insight No.8</h2>
<h3>The highest numbers of dengue cases are in August and September. These months have the most cases and a lot of outlier values compared to others. </h3>

<h2>Insight No.9</h2>
<h3>Based on the graph, the number of dengue cases is generally much lower during the first half of the year, particularly from January to June, compared to the second half, from July to December.</h3>

<h1>6. 
Do dengue deaths happen more often in certain months?</h1>

```python
# Visualization 6: Boxplot of Dengue Deaths by Month
plt.figure(figsize=(8,5))
sns.boxplot(x='Month', y='Dengue_Deaths', data=dengue, palette='pastel')
plt.title('Dengue Deaths Distribution by Month')
plt.xticks(rotation=45)
plt.show()
```

<h2>Insight No.8</h2>
<h3>The data of dengue deaths shows that most deaths occur in September and October. These months have the highest average death counts, and there are also several extreme cases with very high numbers of deaths. </h3>

<h1>7. Is there a correlation between the number of cases and deaths?</h1>

```python
# Visualization 7: Scatter plot of Dengue Cases vs Deaths
plt.figure(figsize=(7,5))
sns.scatterplot(x='Dengue_Cases', y='Dengue_Deaths', hue='Region', data=dengue, palette='tab10')
plt.title('Relationship Between Dengue Cases and Deaths')
plt.show()
```

<h2>Insight No.10</h2>
<h3> There's a positive correlation between the number of dengue cases and deaths. This means that when more people get dengue, more people tend to die from it too.  </h3>

<h2>Insight No.11</h2>
<h3> There are a few regions, such as NCR, that have a relatively small number of dengue cases but a very high number of dengue deaths compared to other regions. </h3>


<h1>8. In which year did dengue cases peak?</h1>

```python
# Visualization 15: Barplot of Total Dengue Cases per Year
plt.figure(figsize=(8,5))
sns.barplot(x='Year', y='Dengue_Cases', data=dengue, estimator=sum, ci=None, color='lightgreen')
plt.title('Total Dengue Cases per Year')
plt.show()
```

<h2>Insight No.12</h2>
<h3> It shows that dengue cases were highest in 2019. The bar for 2019 is the tallest, meaning there were more cases that year than in any other year shown.  </h3>

<h2>Insight No.13</h2>
<h3>Also, it shows that dengue cases were lowest in 2020. And the bar for 2020 is the smallest, which means it has the fewest cases.  </h3>

<h1>9. Which regions and years had the most dengue cases?</h1>

```python
pivot = dengue.pivot_table(index='Region', columns='Year', values='Dengue_Cases', aggfunc='mean'); 
sns.heatmap(pivot, cmap='coolwarm', annot=True, fmt='.0f')
```

<h2>Insight No.14</h2>
<h3>Based on the heatmap, the region and year with the most average dengue cases is Region IV-A in 2019, with an average of 6,350 cases.  </h3>

<h2>Insight No.15</h2>
<h3>Most regions, including NCR and Regions III, IV-A, VI, and IX, experienced their highest number of dengue cases in 2019, and then saw a large decrease in cases the following year, 2020.</h3>
